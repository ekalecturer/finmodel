Although there are technical differences between network theory and graph theory but we will not discuss these differences in this lesson.

Let's look at the undirected network matrix on the left first. The node names are labeled along the top and the left side of the matrix for clarity. In an adjacency matrix, $A_{ij}$ indicates an adjacency matrix with row i and column j. The values are defined as follows:
<br>
<br>
$A_{ij}=1$ if a link pointing from node j to node i exists
<br>
$A_{ij}=0$ otherwise
<br>
<br>
When reading an adjacency matrix, we would start from columns first. For an undirected network's adjacency matrix, the values in column A indicate if there is a link pointing from node A to node B, node C or node D. In the undirected network in figure 4, the links from node A to node B, node C and node D are two-way. Hence the values are 1s. Also in this network, node A does not have a link to itself so the value from node A to node A is 0. Therefore, the values of column A are (0,1,1,1).
<br>
<br>
Looking at column D in the undirected network, the only link that exists is the link pointing from node D to node A. Hence the first entry of the column D is 1. Since there are no other link for node D, the rest of the entries in column D are 0s. Therefore, the values of column D are (1,0,0,0). With the same logic, we can fill in the entries of adjacency matrix for directed network.
<br>
<br>
Some observations about the adjacency matrix:
- The diagonal values of the adjacency matrix for an undirected network are all 0s if every node does not have a link to themselves.
- The adjacency matrix for an undirected network is **symmetric**.
- An adjacency matrix is usually **sparse**. It is because each node would only have a few links to some nodes in the network. That means the adjacency matrix is a sparse matrix.
- When all nodes in a network connect to all other nodes, this network is called a **complete graph**.

The adjacency matrices shown above have entries of only 0s and 1s, representing the presence or absence of connections. These are called **Unweighted Adjacency Matrices**. We can relax this assumption and allow connected nodes to have various strength of connections. When the strength of connections among nodes in a network are not only 1s or 0s, this network has a **Weighted Adjacency Matrix**. Figure 5 shows an example of the weighted adjacency matrix for a directed network.

These coefficients help determine which variables are connected to the dependent variable and the nature of their relationship. The significance of regression parameters plays a crucial role in determining entries in the adjacency matrix. Understanding this relationship between regression coefficients and network connections is essential for properly interpreting regression analysis results in the context of network construction.

## **4. Simple Linear Regression Model for Asset Network**
In this section, we will start by using simple linear regression (OLS) to model the interaction of 6 stock returns (nodes) to construct an asset network.
This asset network will be a bidirectional network, which means the relationship can be reciprocal between any two nodes. For stock i and stock j, we will have one regression that returns for stock i will be the dependent variable and returns of stock j will be the independent variable. But we will also have another regression that returns for stock j will be the dependent variable and returns for stock i will be independent variable. The following regressions demonstrate the two regressions for stock i and stock j.
<br>
<br>
$$R_i = \alpha_{ij} + \beta_{ij}R_j + \varepsilon_i\:\:\:\:\:\:\:\:\:(1)$$
$$R_j = \alpha_{ji} + \beta_{ji}R_i + \varepsilon_j\:\:\:\:\:\:\:\:\:(2)$$
<br>
<br>
Let's look at the regression (1) first. From the standard OLS regression explanation, $\beta_{ij}$ coefficient means when $R_j$ increases by one unit, $R_i$ will increase by $\beta_{ij}$ units. If the estimate of $\beta_{ij}$ coefficient is statistically significant, we can say this regression relationship exists for $R_i$ and $R_j$. In the asset network context, $\beta_{ij}$ shows the tendency that $R_i$ can be influenced by $R_j$. If $\beta_{ij}$ is statistically significant, then the influence exists. We can say that $R_i$ responds to $R_j$, or the link goes from $R_j$ to $R_i$($R_j$ --> $R_i$).Although the above equations look like CAPM, but there are significant differences. In CAPM, the independent variable is a market proxy variable. But here we have an individual stock return as the independent variable.
<br>
<br>
For regression (2), if $\beta_{ji}$ is statistically significant, we can infer that $R_j$ responses to $R_i$. The link between the two nodes goes from $R_i$ to $R_j$($R_i$ --> $R_j$).
<br>
<br>
Our next step is to check p value of $\beta_{ji}$ coefficient for statistical significance. In our simple network model with only two assets, we have two regression models. We have to conduct two tests simultaneously to check for significance of $\beta_{ij}$ and $\beta_{ji}$. Usually we would set our p value threshold at 0.05 for a test. If the p value of $\beta_{ij}$ is less than 0.05, then $\beta_{ij}$ is statistically significantly different from 0. However, we have two tests in this case. We will need to adjust p value to avoid the multiple testing problem.
<br>
<br>
The multiple testing problem arises when conducting several statistical tests simultaneously without adjusting p-values, leading to an inflated risk of false positives. When using the p value threshold of α = 0.05 for a single test, you accept a 5% chance of incorrectly rejecting the null hypothesis. However, this error rate compounds dramatically across multiple tests.
For example, when conducting 5 independent tests at α = 0.05, the probability of obtaining at least one false positive jump to approximately 23% (1 - (1 - 0.05)^5=0.226). There are many methods to adjust p value for multiple tests. In this lesson, we will use Bonferroni method. Please review the required reading to learn more about Bonferroni method. The way Bonferroni method adjusts p value threshold is to divide original threshold (say 0.05) by the number of tests conducted. In the simple two asset network case, the adjusted p value threshold would be 0.05/2=0.025. In our 6-stock returns network, the adjusted p value threshold will be 0.05/30= 0.00167. (Why 30? This question will leave as an exercise for readers to find out).
<br>
<br>
If we confirm that $\beta_{ij}$ is statically significant, we can assert the following link exists: $R_j$ --> $R_i$. We can also fill in the entry in the adjacency matrix using the following rule:
<br>
<br>
$\:\:\:A_{ij} =1\:\:\:\:\:$ if $\beta_{ij}$ is statistically significant
<br>
$\:\:\:A_{ij} =0\:\:\:\:\:$ otherwise
<br>
<br>
With the filled adjacency matrix, we can draw the network and investigate the relationship among asset returns in the asset network. The following Python code implements the steps we describe in this section with the visualization of the asset network.


```python
# Create a dataframe that only includes assets
assets = returns[["Apple","Amazon","GE", "Google","JPMorgan","Warner-Bro","Walmart"]]

# Run a Simple Linear Regression Based Asset Network
n_assets = len(assets.columns)
adj_matrix = np.zeros((n_assets, n_assets))

# Significance threshold (adjusted based on Bonferroni Method)
p_threshold = 0.05/(n_assets*(n_assets-1))

# Loop through all pairs of assets
for i, asset_i in enumerate(assets):
    for j, asset_j in enumerate(assets):
        if i != j:  # Skip self-relationships
            # Run regression: asset_i ~ intercept + asset_j
            model = sm.OLS(assets[asset_i], sm.add_constant(assets[asset_j])).fit()

            # Get p-value for the slope coefficient
            p_value = model.pvalues.iloc[1]

            # If significant, add connection to adjacency matrix
            adj_matrix[i, j] = 1 if p_value < p_threshold else 0

# Create directed graph from adjacency matrix
G = nx.from_numpy_array(adj_matrix, create_using=nx.DiGraph)

# Relabel nodes with asset names
mapping = {i: name for i, name in enumerate(assets)}
G = nx.relabel_nodes(G, mapping)

# Visualize the network
plt.figure(figsize=(10, 8))
pos = nx.spring_layout(G, seed=42)
nx.draw(G, pos,
        node_color='lightblue',
        node_size=500,
        with_labels=True,
        font_size=10,
        edge_color='gray',
        arrows=True,
        arrowsize=15)
plt.title('Simple Linear Regression-Based Asset Network')
plt.show()
```

## **5. Non-Linear Regression Model for Asset Network**
In this section, we test non-linear regression based asset network. We learned that linear regression based asset network does not provide much information about the relationship among 6 stock returns. One way we can adjust the regression model is to add a squared independent variable to make the regression non-linear. Following the two-stock returns example from the last section, the regressions can be adjusted as follows:
<br>
<br>
$$R_i = \alpha_{ij} + \beta_{ij}R_j + \gamma_{ij}R_j^2 + \varepsilon_i\:\:\:\:\:\:\:\:\:(3)$$
$$R_j = \alpha_{ji} + \beta_{ji}R_i +  \gamma_{ji}R_i^2 +\varepsilon_j\:\:\:\:\:\:\:\:\:(4)$$
<br>
<br>
With the adjustment of the regressions, our focus will still be $\beta$ to understand if the linear part of the relationships between stock returns exists.
The following Python code implements the non-linear regression based asset network model for 6 stock returns.

```python
# Run a Non-Linear Regression Based Asset Network
n_assets = len(assets.columns)
adj_matrix = np.zeros((n_assets, n_assets))

# Significance threshold (can be adjusted)
p_threshold = 0.05/(n_assets*(n_assets-1))

# Loop through all pairs of assets
for i, asset_i in enumerate(assets):
    for j, asset_j in enumerate(assets):
        if i != j:  # Skip self-relationships
            # Create squared term
            indep = pd.concat([assets[asset_j],assets[asset_j]**2], axis = 1)
            # Run regression with intercept, original term, and squared term
            X = sm.add_constant(indep)
            model = sm.OLS(assets[asset_i], X).fit()

            # Get p-value for the slope coefficient
            p_value = model.pvalues.iloc[1]

            # If significant, add connection to adjacency matrix
            adj_matrix[i, j] = 1 if p_value < p_threshold else 0

# Create directed graph from adjacency matrix
G = nx.from_numpy_array(adj_matrix, create_using=nx.DiGraph)

# Relabel nodes with asset names
mapping = {i: name for i, name in enumerate(assets)}
G = nx.relabel_nodes(G, mapping)

# Create and visualize the network
plt.figure(figsize=(10, 8))
pos = nx.spring_layout(G, seed=42)
nx.draw(G, pos,
        node_color='lightblue',
        node_size=500,
        with_labels=True,
        font_size=10,
        edge_color='gray',
        arrows=True,
        arrowsize=15)
plt.title('Non-Linear Regression-Based Asset Network')
plt.show()
```

Clearly, using squared returns of stocks does not improve the return of the stock; so this non-linearity is not helping the regression. ## **6. Linear Regression Model with a Confounder for Asset Network**
Now, let's think about our two-stock asset network a little deeper. In the previous sections, we only use the returns of these two stocks to identify their relationship. Sometimes, that model set up may not have enough economic connections. There can be a case where the reason the two stocks are correlated is because they are all influenced by a common factor not in the model. In our case, an equity index may be the common factor influencing both stock returns. Therefore, incorrect regression model specification without well-thought economic connections can produce false variable relationships that actually do not exist.
<br>
<br>
To handle the issue of overfitting or false relationship generated from regression models, researchers have proposed to include factors as confounders in the models (López de Prado, 2003). The key criteria to select a confounder is that the confounder should be correlated to an independent variable(s). The confounder should also causally relate to dependent variable since we are analyzing a directed asset network. Say $R_{C}$ is the return of the confounder variable, we can include $R_{C}$ in our regression models as follows:
<br>
<br>
$$R_i = \alpha_{ij} + \beta_{ij}R_j + \rho_{ij} R_{C}+\varepsilon_i\:\:\:\:\:\:\:\:\:(5)$$
$$R_j = \alpha_{ji} + \beta_{ji}R_i +  \rho_{ji} R_{C} +\varepsilon_j\:\:\:\:\:\:\:\:\:(6)$$
<br>
<br>
The choice of a confounder depends on econometric analysis. In our case, since our assets are all stocks, S&P 500 index is a good option as a confounder as it represents the general stock market exposure.
<br>
<br>
We can rerun our linear regression models with S&P 500 returns and review the result. We are still focusing on $\beta$ estimates from the regression models.

```python
# Run a Linear Regression Based Asset Network with a Confounder
n_assets = len(assets.columns)
adj_matrix = np.zeros((n_assets, n_assets))
SP500 = returns["SP500"]


# Significance threshold (can be adjusted)
p_threshold = 0.05/(n_assets*(n_assets-1))

# Loop through all pairs of assets
for i, asset_i in enumerate(assets):
    for j, asset_j in enumerate(assets):
        if i != j:  # Skip self-relationships
            # Create squared term
            indep = pd.concat([assets[asset_j], SP500], axis = 1)
            # Run regression with squared term
            X = sm.add_constant(indep)
            model = sm.OLS(assets[asset_i], X).fit()

            # Get p-value for the slope coefficient
            p_value = model.pvalues.iloc[1]

            # If significant, add connection to adjacency matrix
            adj_matrix[i, j] = 1 if p_value < p_threshold else 0

# Create directed graph from adjacency matrix
G = nx.from_numpy_array(adj_matrix, create_using=nx.DiGraph)

# Relabel nodes with asset names
mapping = {i: name for i, name in enumerate(assets)}
G = nx.relabel_nodes(G, mapping)

# Create and visualize the network
plt.figure(figsize=(10, 8))
pos = nx.spring_layout(G, seed=42)
nx.draw(G, pos,
        node_color='lightblue',
        node_size=500,
        with_labels=True,
        font_size=10,
        edge_color='gray',
        arrows=True,
        arrowsize=15)
plt.title('Linear Regression-Based Asset Network with S&P500 as Confounder')
plt.show()
```

By controlling the confounder variable in our regression models, the new asset network reveals some interesting information. From the graph, we can see that JPMorgan and GE the only two which have bidirectional relationships to all other stocks. JPMorgan and GE can be treated as two hubs in this asset network. We all know that JPMorgan is a financial institution so it should have business relationship with other companies. As for GE, it also has a financing division in its business and can be treated as a financial institution too. The next interesting observation is all three technology stocks (Apple, Amazon and Google) have bidirectional relationship too. It shows stocks in technology section can be highly interconnected. The last interesting observation to point out is that Warner Brothers stock return and Walmart's stock returns are not connected after including S&P 500 index return to the model. It says that both stock returns are not directly correlated after taking into account of S&P500 index return. We say Warner Brothers stock returns and Walmart's stock returns are **conditionally independent**. The conditionally independence also happens to other pairs.
<br>
<br>

## **7. Asset and Factor LASSO Model**
#### ***7.1 Overview***
The previous section demonstrated the power of adding a confounder variable to an asset network. Now we face a new challenge: what if multiple potential confounders exist, all with economic justification? We need a more efficient method to test these confounders in our regression models when estimating network structure. In this section, we'll refer to these confounder variables as **Factors** and introduce **Graphical LASSO** (Graphical Least Absolute Shrinkage and Selection Operator).
<br>
<br>
Our goal remains to identify direct relationships between assets while controlling other factors. These direct relationships form the foundation of our asset and factor network. **Partial correlation** is the appropriate measure for evaluating these direct relationships. Unlike standard correlation, partial correlation quantifies the relationship between two variables while controlling the effects of other variables. This approach helps identify genuinely direct connections between assets, filtering out indirect relationships that occur through intermediary assets, thus providing a clearer picture of the underlying network structure in financial markets. Please review the required reading "Conditional Distributions" for more information about partial correlation.
<br>
<br>
In the network context, we will need to have a partial correlation matrix to describe partial correlations for all assets and factors in the network. From there we can build links among nodes. In the real financial market, not all assets and factors are all partially correlated. Many of the partial correlations can be 0s. To translate this observation to partial correlation matrix, we can say partial correlation matrix is **sparse**. In order to identify sparsity in the partial correlation matrix, we can use LASSO or L1 penalty to impose constraints while estimating partial correlation matrix. The lasso penalty enforces this sparsity by shrinking many partial correlations exactly to zero. When the partial correlation of two variables 0 we say these two variables are conditionally independent.
<br>
<br>
#### ***7.2 Graphical LASSO Model***
In a Graphic LASSO model, there are $p$ random variables in a dataset $\textbf{X}$ follows a multivariate normal distribution as follows:
 $$\textbf{X} = (X_{1}, X_{2},...,X_{p})\sim MVP(\mu,\Sigma)$$
<br>
<br>
Let's define $\Theta=\Sigma^{-1}$, where $\Theta$ is the inverse of the covariance matrix. $\Theta$ is called **Precision Matrix**. The elements of this precision matrix are proportional to the partial correlation matrix. By estimating the precision matrix, we can determine the partial correlations needed to construct our network.
<br>
<br>
To incorporate the sparse nature of partial correlation matrices in asset and factor networks, we apply a LASSO penalty to the model. Friedman et al. (2008) proposed an efficient method for solving the inverse covariance matrix in a Graphical LASSO model by maximizing the following penalized log-likelihood function with respect to the mean vector $\mu$:
<br>
<br>
$$log\text{ }det(\Theta)-trace(S\Theta)-\rho\left\| \Theta \right\|_{1}$$
<br>
where
<br>
$\Theta$ is the inverse of covariance matrix
<br>
$S$ is the sample covariance matrix
<br>
$\rho$ is the penalty parameter to control sparsity
<br>
$\left\| \Theta \right\|_{1}$ is the L1 norm—the sum of the absolute values of the elements of $\Theta$
<br>
<br>
The parameter $\rho$ is crucial: increasing $\rho$ produces a sparser network by forcing more elements of $\Theta$ to zero, which in turn makes the corresponding partial correlations zero. Selecting an appropriate value for $\rho$ requires careful consideration. If $\rho$ is too high, we'll have no connections among assets; if $\rho$ is too low, all assets will be connected, rendering the network uninformative. In our application, we'll employ cross-validation to determine the optimal $\rho$.
<br>
<br>
#### ***7.3 Application***
In this section, we'll apply the Graphical LASSO model to the dataset described in Section 3.2 to construct an asset and factor network. We'll utilize all 7 stock returns we used in the previous sections. We will also add S&P500 ETF, Treasury ETF, Financial sector ETF, Technology ETF and US Dollar Index as factors to the model. This model will demonstrate how the precision matrix can reveal the underlying structure of financial markets while controlling common factors.
<br>
<br>
First, let's standardize all the stock and factor return data.

```python
# Standardize data
standardized_returns = (returns - returns.mean()) / returns.std()
```

Next, we will use cross validation method to find the optimal level of  $\rho$  for analysis.

```python
# Create a range of rhos to test
rhos = np.linspace(0.3, 1, 100)

# Use GraphicalLassoCV to find the best alpha
model_cv = GraphicalLassoCV(alphas=rhos, cv=5, tol=1e-2, max_iter=400)
model_cv.fit(standardized_returns)

#best_rho = 0.3
best_rho = model_cv.alpha_
print(f"Best ρ: {best_rho}")
```

With the optimal $\rho$ available, we can now fit the graphical LASSO model and estimate the inverse of covariance matrix. Then we can calculate partial correlation matrix.

```python
# Fit the model with the best alpha
model = GraphicalLasso(alpha=best_rho, max_iter=400, tol=1e-2)
model.fit(standardized_returns)

# Extract precision matrix (inverse covariance)
precision_matrix = model.precision_

# Convert precision matrix to partial correlation matrix for easier interpretation
d = np.sqrt(np.diag(precision_matrix))
partial_corr = -precision_matrix / np.outer(d, d)
np.fill_diagonal(partial_corr, 0)  # Remove self-loops
partial_corr
```

Once we obtain the partial correlation matrix, we will use it to generate the adjacency matrix. We use the rule that if the absolute value of the partial correlation between to nodes are greater than 0, the entry for the adjacency matrix will 1. Otherwise the entry will be 0.

```python
# Create adjacency matrix using a threshold
threshold = 0.0  # Adjust as needed
adjacency = (np.abs(partial_corr) > threshold).astype(int)
adjacency
```

Now, we are ready to draw the network!

```python
# Create network
G = nx.from_numpy_array(adjacency)

# Relabel nodes with asset and factor names
mapping = {i: name for i, name in enumerate(returns.columns)}
G = nx.relabel_nodes(G, mapping)

# Assign types to nodes
node_types = {
    'Apple':'Asset',
    'Amazon':'Asset',
    'USD':'Factor',
    'GE':'Asset',
    'Google':'Asset',
    'JPMorgan':'Asset',
    'Treasury':'Factor',
    'Warner-Bro':'Asset',
    'Walmart':'Asset',
    'Financial':'Factor',
    'Technology':'Factor',
    'SP500':'Factor',
}

nx.set_node_attributes(G, node_types, 'type')

# Draw the network graph
plt.figure(figsize=(14, 10))

# Use different colors for assets and factors
node_colors = ['skyblue' if G.nodes[name]['type'] == 'Asset' else 'orange' for name in G.nodes()]

# Create layout
pos = nx.spring_layout(G, seed=42)

# Draw the network
nx.draw(G, pos,
        with_labels=True,
        node_size=500,
        node_color=node_colors,
        font_size=10,
        font_weight='bold',
        edge_color='gray',
        arrows=True,
        arrowsize=15)

# Add a legend
asset_patch = plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='skyblue',
                       markersize=15, label='Asset')
factor_patch = plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='orange',
                        markersize=15, label='Factor')
plt.legend(handles=[asset_patch, factor_patch], loc='upper right')

plt.title('Factor-Asset Network via Graphical LASSO', fontsize=16)
plt.show()
```

From the above asset and factor network graph, we have some interesting observations.
<br>
1. S&P500 ETF, Financial ETF and Technology ETF take up the central section of the network, representing their high connections with other stocks.
2. Treasury EFT and USD index only connect to each other and forms a separate cluster, demonstrating their lack of connection to the 7 stocks
3. Three technology stock returns (Apple, Amazon, and Google) form a cluster and are connected to each other. They are also connected to technology ETF and S&P500 index.
4. JP Morgan, GE and financial ETF all connect to each other and form a financial cluster.
5. Walmart is only connected to S&P500 index.

With the above observations, we can see the 3 technology stocks are correlated. It can be an issue if all 3 stocks have significant allocation in one portfolio. The portfolio manager may have to review the asset allocation and adjust the holdings of these 3 technology stocks. Another insight we can get from the network graph is that treasury ETF and USD are not correlated to the equity markets. They can be good alternative investment options to add to an all equity portfolio to diversity in equity market risk.

#### ***7.4 Network Comparison with Correlation Matrix and PCA***
We analyzed the result of the asset factor network in the last section. While the Graphical LASSO-based network reveals conditional dependencies between assets and factors, it is interesting compare these findings with traditional correlation analysis and Principal Component Analysis (PCA). Let's start with the correlation analysis. We will create and correlation matrix and draw a heatmap for analysis.

```python
# Calculate the correlation matrix for standardized stock returns and draw a correlation heatmap
corr_matrix = standardized_returns.corr()

# Create the heatmap
plt.figure(figsize=(10, 8))  # Adjust figure size as needed
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Standardized Stock Returns Correlation Heatmap')
plt.show()
```

From the above correlation heatmap, we can see similar conclusions we obtained from the asset and factor network. What is interesting is the the asset and factor network provides more sparse connections in the Graphical LASSO network. In this heatmap, we see decent correlations between Apple and Financial ETF, JPMorgan and Technology ETF, and Google and Financial ETF pairs. However, we don't see the links in the asset factor network. This suggests that the additional high correlation pairs from the heatmap are mediated through other variables in the asset factor network.
<br>
<br>
Next, let's check PCA analysis.

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=12)
pca.fit(standardized_returns)

# Get proportions of variance and cumulative proportion of variance
pr_var = pca.explained_variance_ratio_
cum_pr = np.cumsum(pca.explained_variance_ratio_)
ind = ["Proportion of variance", "Cumulative proportion of variance"]
cols = ["PC1", "PC2", "PC3", "PC4", "PC5", "PC6", "PC7", "PC8", "PC9", "PC10", "PC11", "PC12"]
pd.DataFrame(np.vstack((pr_var, cum_pr)), ind, columns=cols)
```

From the above PCA result, we can see that the first component takes up 47% of data variation. The first 3 components account for almost 70% of the data variation. It means that there are concentrated factors for our dataset. Now let's check the loadings.

```python
# Coefficients (Loadings) of 11 Principal Components
pc_res = pd.DataFrame(pca.components_.T, index=list(standardized_returns.columns), columns=cols)
pc_res
```

The above loading table gives us some key insights:

1. The key elements in PC1 are S&P500 ETF, Technology ETF and Financial ETF. It represents the overall equity market factor
2. The key elements in PC2 are USD index and Treasury ETF. This component captures the similar cluster of USD index and Treasury ETF cluster in the asset and factor network
3. PC3 seems to be the component that have negative impact on Financial sector but positive impact on technology sector.

Overall, the result from PAC is complement to the result from asset factor network. The network analysis tells us the connections and among assets and factors while PCA provides insights what are the possible common drivers that form these dependencies.
